# Columns Operations and More 

We are going to use IMDB Top-250 movies dataset in this example.
You can download the dataset with link in the below.

https://www.kaggle.com/datasets/yehorkorzh/imdb-top-250-movies



```python
# 1-Import the pandas library and alias it as pd.
import pandas as pd
```

```python
# 2-Import the dataset and assign it to a variable named films.

films = pd.read_csv("imdb_movies.csv")
films.head()
```

```python
# 3-What are the names of the columns?

films.columns
```

```python
# 4-Set the 'Num' column as the index of the dataframe

films.set_index("Num",inplace=True)
```

```python
#Â 5-Drop the "Unnamed: 0" column from the dataset.

films.drop(columns="Unnamed: 0",inplace=True)
```

```python
# 6-Print only the column Title

films.Title
```

```python
# 7-Print the columns 'Title' and 'Genre'.

films[["Title","Genre"]]
```

```python
# 8-Select the 'Title', 'Runtime' and 'Genre' columns from row 1-2 and 3 .

films.loc[[1,2,3],["Title","Runtime","Genre"]]
```

```python
# 9- Select the rows 3 to 7 and the columns 3 to 6

films.iloc[2:7,2:6]
```

```python
# 10-Select every row after the fourth row and all columns

films.iloc[4:,:]
```

```python
# 11-Select rows where film.Runtime is greater than 140.

# First we have to change our Runtime column type as integer after that we can use ">" operators.

films.Runtime.astype(int)
```

```python
# We have to clean our data because while we try to change our data type error occur we need to fix it.

films.Runtime = films.Runtime.str.strip(" min")
```

```python
films.Runtime = films.Runtime.astype(int)
```

```python
films[films.Runtime>140]
```

```python
# 12-Rename Runtime column and make it understandable (Runtime (min)).

films.rename(columns={"Runtime":"Runtime (min)"},inplace=True)
films[["Title","Runtime (min)"]]
```

```python
# 13-Select rows where Runtime is greater than 210 or less than 80

filt = (films["Runtime (min)"] > 210) | (films["Runtime (min)"]<80)
films[filt]
```
